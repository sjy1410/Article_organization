```python
#모듈 추가
import random
import pandas as pd
```

```python
#조번호와 4개의 반 번호를 저장하는 리스트 생성
c1=[]
c2=[]
c3=[]
c4=[]

i=0
for i in range(1, 17) :
  c1.append(i)

for i in range(1, 17) :
  c2.append(i)

for i in range(1, 17) :
  c3.append(i)

for i in range(1, 17) :
  c4.append(i)
```

```python
#각반 번호 섞기
random.shuffle(c1)
random.shuffle(c2)
random.shuffle(c3)
random.shuffle(c4)
```

```python
#판다스 데이터프레임
lt=[]
lt.append(list(range(1, 17)))
lt.append(c1)
lt.append(c2)
lt.append(c3)
lt.append(c4)

df = pd.DataFrame(lt)
df
```

```python
#행과 열을 전환
df = df.transpose()
df
```

```python
#열인덱스 설정
df.columns=['조', '1반', '2반', '3반', '4반']
df
```

```python
#조번호를 행인덱스로 설정
df=df.set_index('조')
df
```

```python
#예외처리
df.loc[df[df['1반']==16].index, '1반']='없음'
df.loc[df[df['2반']==16].index, '2반']='빠질수있음'
df.loc[df[df['4반']==16].index, '4반']='없음'
df.loc[df[df['4반']==16].index, '4반']=17
df
```
